```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('churn2.xlsx')
data.head()
```

```python
data=data.drop('code',axis=1)
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
li=['city','churn','namava','shoteltalk','casper','Android Security']
data[li]=data[li].astype(str)
data.info()
```

```python
sp=data['speed']
sp_n=(sp-sp.mean(axis=0))/(sp.std(axis=0))
data['speed']=sp_n
data
```

```python
x=data.drop('churn',axis=1)
y=data['churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
from sklearn.naive_bayes import GaussianNB
gs=GaussianNB()
gs.fit(x_train,y_train)
y_pred=gs.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from IPython.display import display
from sklearn.metrics import classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
adb=AdaBoostClassifier(GaussianNB(),n_estimators=20,random_state=101)
adb.fit(x_train,y_train)
ypred=adb.predict(x_test)
ypred
```

```python
accuracy_score(y_test,ypred)
```

```python
confusion_matrix(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```
